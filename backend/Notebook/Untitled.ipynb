```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=30)
```

```python
X_test.shape,X_train.shape,y_test.shape,y_train.shape
```

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
 models = {
    'RandomForest': RandomForestClassifier(random_state=30),
    'GradientBoosting': GradientBoostingClassifier(random_state=30)
}
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
for model_name,model in models.items():
    model.fit(X_train,y_train)
    prediction=model.predict(X_test)
    score=accuracy_score(y_test,prediction)
    cm=confusion_matrix(y_test,prediction)
    print(prediction)
    print('score :',score)
    print('matrix :', cm)
    
```

```python

def helper(dis):
    descr= description[description['Disease']==dis]['Description']
    descr=" ".join([w for w in descr])
    prec=precuation[precuation['Disease']==dis][['Precaution_1','Precaution_2','Precaution_3','Precaution_4']];
    prec=[col for col in prec.values]
    med=medication[medication['Disease']==dis]['Medication']
    med= [med for med in med.values]
    die=diet[diet['Disease']==dis]['Diet']
    die=[col for col in die.values]
    work=workout[workout['disease']==dis]['workout']
    return descr,prec,med,die,work
```
